```python
class Cliente:
    '''
    Clase que crea objetos de tipo cliente
    '''
    def __init__(self, nombre: str, apellido: str, genero: str, ocupacion: str):
        self._nombre = nombre
        self._apellido = apellido
        self._genero = genero
        self._ocupacion = ocupacion

    # --- Métodos que permiten conocer (getter) los atributos ---

    @property
    def nombre(self):
        return self._nombre

    @property
    def apellido(self):
        return self._apellido

    @property
    def genero(self):
        return self._genero

    @property
    def ocupacion(self):
        return self._ocupacion

    # --- Métodos que permiten modificar (setter) los atributos ---

    @nombre.setter
    def nombre(self, nuevo_nombre: str):
        self._nombre = nuevo_nombre

    @apellido.setter
    def apellido(self, nuevo_apellido: str):
        self._apellido = nuevo_apellido

    @genero.setter
    def genero(self, nuevo_genero: str):
        self._genero = nuevo_genero

    @ocupacion.setter
    def ocupacion(self, nueva_ocupacion: str):
        self._ocupacion = nueva_ocupacion

    def __str__(self):
        return f'Cliente [{self.__dict__.__str__()}]'


# --- Programa principal ---
if __name__ == '__main__':
    cliente1 = Cliente(
        nombre="Maria",
        apellido="Paz",
        genero="Femenino",
        ocupacion="Ingeniera"
    )
    print(cliente1)

    # Modificar un atributo usando el setter
    cliente1.nombre = "Ana"
    print(cliente1)
```
