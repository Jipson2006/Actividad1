```python
class Empleado:
    def __init__(self, nombre, apellido, cedula, email, semestre, materias):
        self.nombre = nombre
        self.apellido = apellido
        self.cedula = cedula
        self.email = email
        self.semestre = semestre
        self.materias = materias

    # --- Encapsulamiento con Decoradores ---
    @property
    def nombre(self): return self._nombre
    @nombre.setter
    def nombre(self, v): self._nombre = v

    @property
    def apellido(self): return self._apellido
    @apellido.setter
    def apellido(self, v): self._apellido = v

    @property
    def cedula(self): return self._cedula
    @cedula.setter
    def cedula(self, v):
        if len(str(v)) == 10 and str(v).isdigit():
            self._cedula = v
        else:
            raise ValueError("Cédula inválida: debe tener 10 dígitos.")

    @property
    def email(self): return self._email
    @email.setter
    def email(self, v): self._email = v

    @property
    def semestre(self): return self._semestre
    @semestre.setter
    def semestre(self, v):
        if 1 <= int(v) <= 9:
            self._semestre = v
        else:
            raise ValueError("Semestre inválido: debe ser entre 1 y 9.")

    @property
    def materias(self): return self._materias
    @materias.setter
    def materias(self, v): self._materias = v

    # MÉTODO PARA MOSTRAR LA INFORMACIÓN COMPLETA
    def __str__(self):
        return (f"--- Ficha del Empleado ---\n"
                f"Nombre Completo: {self.nombre} {self.apellido}\n"
                f"Cédula: {self.cedula}\n"
                f"Email: {self.email}\n"
                f"Semestre: {self.semestre}\n"
                f"Materias: {', '.join(self.materias)}\n"
                f"--------------------------")

# --- PRUEBA DEL PROGRAMA ---
try:
    # Creamos el objeto con todos los datos
    empleado1 = Empleado(
        nombre="Jipson",
        apellido="Silva",
        cedula="0932280274",
        email="Jipson.s@empresa.com",
        semestre=3,
        materias=["Cálculo", "Programación OO", "Física"]
    )

    print(empleado1)

except ValueError as e:
    print(f"Error en los datos: {e}")
```
